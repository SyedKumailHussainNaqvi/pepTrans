```python
import numpy as np
import random

np.random.seed(42)
random.seed(42)

rmse_per_fold = []
pearson_per_fold = []

for fold_no in range(1, 6):
    print(f"\nTraining on Fold {fold_no}")

    X_train, y_train = train_data_dict[f"fold_{fold_no}"]
    X_test, y_test = test_data_dict[f"fold_{fold_no}"]

    with tf.device('/GPU:1'):
        model = pepTrans_Binding_Affinity(2048, X_train)
        early_stopping = EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True)

        history = model.fit(
            X_train, y_train,
            epochs=2000,
            batch_size=128,
            validation_data=(X_test, y_test),
            callbacks=[early_stopping],
            verbose=1
        )

    scores = model.evaluate(X_test, y_test, verbose=0)
    predictions = model.predict(X_test).flatten()
    test_labels = y_test.flatten()

    rmse = mean_squared_error(test_labels, predictions, squared=False)
    pearson_corr, _ = pearsonr(test_labels, predictions)

    rmse_per_fold.append(rmse)
    pearson_per_fold.append(pearson_corr)

    print(f"Fold {fold_no} - Loss: {scores[0]:.4f}, RMSE: {rmse:.4f}, Pearson: {pearson_corr:.4f}")

    train_loss = history.history['loss']
    val_loss = history.history['val_loss']
    plt.figure(figsize=(5, 2))
    plt.plot(train_loss, label='Training Loss')
    plt.plot(val_loss, label='Validation Loss')
    plt.axvline(np.argmin(val_loss), color='red', linestyle='--', label='Early Stop')
    plt.title(f'Fold {fold_no} Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

print("\n" + "="*40)
print("Cross-Validation Performance Summary")
print("="*40)
for i in range(5):
    print(f"Fold {i+1}: RMSE = {rmse_per_fold[i]:.4f}, Pearson = {pearson_per_fold[i]:.4f}")
print("-"*40)
print(f"Mean RMSE: {np.mean(rmse_per_fold):.4f} ± {np.std(rmse_per_fold):.4f}")
print(f"Mean Pearson: {np.mean(pearson_per_fold):.4f} ± {np.std(pearson_per_fold):.4f}")
```
